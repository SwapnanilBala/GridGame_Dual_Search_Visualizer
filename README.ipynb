## 1 &ensp; Overview

An interactive pathfinding toolkit that lets us **run two search algorithms simultaneously** on the same grid map, visualize their exploration in real time with **Pygame**, and generate comparative analytics with **Matplotlib**.

Each algorithm searches for a path from **S** (start) to **G** (goal) through a 2D grid that may contain walls, open floor, and weighted terrain tiles.

**Key features:**
- Side-by-side dual algorithm viewer with synchronized playback
- Animated node expansion, frontier highlighting, and path tracing
- Automatic comparison chart generation (bar charts + summary table)
- Terrain-aware costs &mdash; Road (1), Mud (5), Water (10)
- Interactive launcher UI for selecting maps, algorithms, and animation speed

---
## How the Algorithms Differ

All six algorithms solve the same problem — find a path from **S (Source/ RED BOX)** to **G (Goal/ GREEN BOX)** — but they differ fundamentally in **how they choose which node to explore next**, which directly affects optimality, efficiency, and memory usage.

| Aspect | Details |
|:-------|:--------|
| **Uninformed vs Informed** | BFS, DFS, DLS, UCS, and BDS are *uninformed* — they have no knowledge of where the goal lies. A\* is *informed* because it uses a heuristic (Manhattan distance) to estimate proximity to the goal, steering the search toward it. |
| **Cost-blind vs Cost-aware** | BFS, DFS, DLS, and BDS treat every step as equal cost. UCS and A\* account for actual terrain costs (Mud = 5, Water = 10), finding the *cheapest* path rather than simply the shortest by hop count. |
| **Data structure = behaviour** | BFS uses a **FIFO queue** → explores evenly layer by layer. DFS uses a **LIFO stack** → dives deep before backtracking. UCS and A\* use a **priority queue** → always expand the most promising node first. |
| **Optimality trade-offs** | BFS and BDS guarantee the shortest path when all step costs are equal. UCS and A\* guarantee the cheapest path on weighted maps. DFS and DLS offer **no optimality guarantee** — they may reach the goal through a long, expensive detour. |
| **Bidirectional advantage** | BDS runs two BFS frontiers simultaneously — one from start, one from goal — and stops when they meet, roughly halving the explored space compared to single-direction BFS. |

## 2 &ensp; Implemented Algorithms

| # | Algorithm | Module | Strategy | Optimal? | Notes |
|:-:|:----------|:-------|:---------|:--------:|:------|
| 1 | **A\*** | `algos/search/astar.py` | Best-first with `f = g + h` | Yes | Manhattan distance heuristic; requires admissible `h` and non-negative costs |
| 2 | **BFS** | `algos/search/bfs.py` | FIFO queue | Yes | Optimal when all step costs are equal; explores layer by layer |
| 3 | **DFS** | `algos/search/dfs.py` | LIFO stack | No | Memory-efficient but may find suboptimal, winding paths |
| 4 | **DLS** | `algos/search/dls.py` | Depth-limited DFS | No | DFS with a configurable maximum depth cap |
| 5 | **UCS** | `algos/search/ucs.py` | Priority queue on path cost `g` | Yes | Handles weighted terrain; generalizes BFS to non-uniform costs |
| 6 | **BDS** | `algos/search/bds.py` | Dual frontier (start + goal) | Yes | Bidirectional search &mdash; explores from both ends simultaneously |

## 5 &ensp; Getting Started

### 5.1 &ensp; Prerequisites

- **Python 3.12+**

### 5.2 &ensp; Install Dependencies

```bash
pip install -r requirements.txt
```

This installs:
- `matplotlib >= 3.8` &mdash; comparison chart generation
- `pygame >= 2.6.1` &mdash; real-time grid visualization

### 5.3 &ensp; Run the Launcher (Dual Viewer)

```bash
python main.py
```

This opens the interactive Pygame launcher:

| Step | Keys | Action |
|:----:|:-----|:-------|
| 1 | `LEFT` / `RIGHT` | Select a map |
| 2 | `A` / `D` | Choose Algorithm 1 |
| 3 | `J` / `L` | Choose Algorithm 2 |
| 4 | `UP` / `DOWN` | Adjust animation FPS (5&ndash;120) |
| 5 | `ENTER` | Run both algorithms side-by-side |

### 5.4 &ensp; Run a Single Algorithm

```bash
python main.py --mode single --map A-Star.txt --algo astar
```

Available `--algo` values: `astar`, `bfs`, `dfs`, `dls`, `ucs`, `bds`

### 7.1 &ensp; A\* vs BFS &mdash; `A-Star.txt`

| Metric | A\* | BFS |
|:-------|----:|----:|
| Path Cost | 30 | 30 |
| Path Length | 31 | 31 |
| Nodes Expanded | **65** | 109 |
| Max Frontier Size | 8 | **7** |

Both algorithms find **the same optimal path** with identical cost and length, as expected on a uniform-cost map where every step costs 1. The key difference is in **nodes expanded**: A\* expands 65 nodes compared to BFS's 109, a ~40% reduction. This is because A\*'s Manhattan distance heuristic directs the search toward the goal, pruning branches that lead away from it. BFS, being uninformed, has no choice but to explore every cell at a given distance before moving further. The frontier sizes are comparable (8 vs 7), indicating both algorithms use similar memory.

### 7.2 &ensp; DFS vs BFS &mdash; `BFS_Alternate.txt`

| Metric | DFS | BFS |
|:-------|----:|----:|
| Path Cost | 82 | **34** |
| Path Length | 83 | **35** |
| Nodes Expanded | **117** | 137 |
| Max Frontier Size | 15 | **8** |

This comparison highlights the **non-optimality of DFS**. BFS finds a path of length 35 (cost 34), while DFS produces a path more than **twice as long** (83 steps, cost 82). DFS uses a stack and dives deep into the maze's dead ends before backtracking. It happens to find *a* path, but that path winds through the maze inefficiently. BFS, by exploring layer by layer, guarantees it finds the shortest hop-count path. Interestingly, DFS expands fewer nodes (117 vs 137) — it stumbled onto a path before exhausting the entire maze — but that "efficiency" came at the cost of path quality.

### 7.3 &ensp; UCS vs A\* &mdash; `UCS_Weighted.txt`

| Metric | UCS | A\* |
|:-------|----:|----:|
| Path Cost | 106 | 106 |
| Path Length | 39 | 39 |
| Nodes Expanded | 141 | **138** |
| Max Frontier Size | 10 | **9** |

On a map with **Mud (cost 5)** and **Water (cost 10)** tiles, both UCS and A\* now find the **same optimal weighted path** (cost = 106, length = 39). After updating A\* to account for actual terrain costs in its path cost calculation, it no longer treats all steps as equal — it properly incorporates edge weights into `g(n)` while still using the Manhattan heuristic for `h(n)`. A\* expands **138 nodes** versus UCS's **141** — a modest ~2% reduction — because the heuristic provides only a slight advantage on this weighted map. Both algorithms use comparable memory (frontier sizes of 9 vs 10). This matchup confirms that a cost-aware A\* with an admissible heuristic finds the same cheapest path as UCS, while expanding marginally fewer nodes thanks to heuristic guidance.

### 7.4 &ensp; DLS vs DFS &mdash; `DLS.txt`

| Metric | DLS | DFS |
|:-------|----:|----:|
| Path Cost | 50 | 50 |
| Path Length | 51 | 51 |
| Nodes Expanded | **226** | 301 |
| Max Frontier Size | 2 | 2 |

Both algorithms find the same path (cost = 50, length = 51), but DLS expands **226 nodes** compared to DFS's **301** — a 25% reduction. This map was designed with a massive zigzag dead-end corridor (~250 steps deep) that DFS enters first due to its LIFO stack ordering. DFS explores the entire dead end before backtracking to find the real path. DLS, with its depth limit of 175, **cuts off** exploration partway through that dead end, saving 75 node expansions. This is exactly the scenario where depth-limited search shines: preventing the algorithm from wasting time in deep, fruitless branches.

### 7.5 &ensp; BDS vs BFS &mdash; `BDS.txt`

| Metric | BDS | BFS |
|:-------|----:|----:|
| Path Cost | 30 | 30 |
| Path Length | 31 | 31 |
| Nodes Expanded | **100** | 141 |
| Max Frontier Size | 12 | **9** |

Both algorithms find the optimal path (cost = 30, length = 31), confirming that BDS preserves BFS's optimality guarantee on uniform-cost maps. BDS expands **100 nodes** versus BFS's **141**, a ~29% reduction. The advantage is that BDS searches from both the start and the goal simultaneously, meeting in the middle — the explored volume grows as roughly two smaller spheres rather than one large one. The trade-off is a **larger peak frontier** (12 vs 9) because BDS maintains two separate queues that can accumulate more boundary nodes. On this map the savings are substantial, demonstrating how bidirectional search can meaningfully cut down exploration.

## 8 &ensp; Metrics Tracked

Every search run produces a `SearchResult` dataclass:

```python
@dataclass(slots=True)
class SearchResult:
    found: bool
    path: list[State]
    cost: float
    expanded: int
    maximum_nodes_waiting_to_be_explored: int
```

| Metric | Description |
|:-------|:------------|
| `found` | Whether a path from **S** to **G** was discovered |
| `path` | Ordered list of `(row, col)` states from start to goal |
| `cost` | Total movement cost along the path (terrain-weighted) |
| `expanded` | Number of nodes popped from the frontier during search |
| `maximum_nodes_waiting_to_be_explored` | Peak frontier size &mdash; measures memory pressure |
